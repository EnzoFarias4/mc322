# Números Primos

```c
#include <stdio.h>

int main() {
    int numero = 7,
        flag=0;

    for (int i = 2; i < numero; i++)
        if (numero % i == 0)
            flag = 1;

    if (flag == 0)
        printf("Este numero eh primo");
    else
        printf("Não eh primo");


    return 0;
}
```

```c
#include <stdio.h>

int main() {
    int n = 10,
        flag;

    int numero = 1;
    int t = 0;
    while (t < n) {
        flag = 0;
        for (int i = 2; i < numero; i++)
            if (numero % i == 0)
                flag = 1;

        if (flag == 0) {
            printf("%d\n", numero);
            t++;
        }
        
        numero++;
    }

    return 0;
}
```

# Exercício

Construa um gerador de números primos seguindo a lógica de um Tipo Abstrato de Dados.

```c
#include <stdio.h>
#include <stdlib.h>
typedef struct node{int key;struct node *next;} node;
typedef struct lista{node *h;node *t;} lista;
typedef struct Primos{lista *P;int N;} Primos;

lista *init(void)
{
    lista *l = malloc(sizeof(lista));
    l->h = l->t = NULL;
    return l;
}

void enqueue(lista *l,int k)
{
    node *x = malloc(sizeof(node));
    x->key = k;
    x->next = NULL;
    if(!l->h){l->h = x;l->t = x;return ;}
    l->t->next = x;
    l->t = x;
}

int is_prime(int x)
{
       for(int i = 2;x >= i*i; i++)
       {
           if(!(x % i)){return 0;}
       }
    return 1;
}

Primos *Gerar(int N)
{
       Primos *p = malloc(sizeof(Primos));
       p->P = init();
       p->N = 0;
       int num = 2;
       while(N > p->N) 
       {
            if(is_prime(num))
            {enqueue(p->P, num);p->N++;}
            num++;
       }
       return p;
}

void freelist(node *x)
{
    if(x)
    {
        freelist(x->next);
    }
    free(x);  
}

int main(void)
{
    Primos *h;
    h = Gerar(10);
    node *x = h->P->h;
    while(x)
    {
     printf("%d\n",x->key);
     x = x->next;
    }
    freelist(h->P->h);
    free(h->P);
    free(h);
    return 0;
}
```
